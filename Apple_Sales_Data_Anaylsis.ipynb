# What are the top 10 highest rated iPhones on Flipkart in India?

```python
data.head()
```

```python
highest_rating = data.sort_values(by = ['Star Rating'], ascending = False)

highest_rating = highest_rating.head(10)
```

```python
print(highest_rating['Product Name'])
```

```python
print (highest_rating)
```

**The Apple iPhone 11 Pro Max (Gold, 64 GB) is the most highest rated.**

# How many ratings do the highest rated iPhones on Flipkart have?

```python
data.head()
```

```python
iPhones = highest_rating['Product Name'].value_counts()
```

```python
label = iPhones.index
counts = highest_rating['Number Of Ratings']
figure = px.bar(highest_rating, x=label, y=counts)
figure.show()
```

**The Apple iPhone XR Coral has the highest rating of 79.512k**

# Which iPhone has the highest number of reviews on Flipkart?

```python
iPhones = highest_rating['Product Name'].value_counts()
label = iPhones.index
counts = highest_rating['Number Of Reviews']
figure = px.bar(highest_rating, x=label, y=counts, title='Number of reviews of Highest Rated iPhones')
figure.show()
```

**The Apple iPhone XR has the highest number of Reviews (6796)**

# What is the relationship between the sale price of iPhones and the number of ratings on Flipkart?

```python
data.head()
```

```python
figure = px.scatter(data_frame = data, x="Number Of Ratings", y="Sale Price", size="Discount Percentage",trendline="ols",title = "Relationship between sales price and number of ratings")

figure.show()
```

**There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means the iPhones with lower sale prices are sold more in India.**

# What is the relationship between the discount percentage and the number of ratings of iPhones on Flipkart?

```python
figure = px.scatter(data_frame = data,
                    x="Number Of Ratings",
                    y="Discount Percentage",
                    size="Sale Price",
                    trendline="ols",
                    title = "Relationship between discounts and number of ratings")

figure.show()
```

**The relationship is directly proportional. If Discount Percentage increases, the number of Ratings also increases.**

# Can you figure out the least expensive and most expensive iPhones in the Indian market, along with all their specifications?

```python
data.head()
```

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

print ("Most Expensive Product:")
print (most_expensive)

print ("Least Expensive Product:")
print (least_expensive)
```

**The most expensive product is the APPLE iPhone 12 Pro (Silver, 512 GB).
The least expensive product is the APPLE iPhone SE (White, 64 GB)**
